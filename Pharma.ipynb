```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
df = pd.read_csv('pharma-data.csv')
df.head()
```

```python
df.shape
```

```python
df.info
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df1 = df.copy
```

```python
df1 = df[(df['Quantity'] >= 0) & (df['Price'] >= 0)]
```

```python
df1['calc_amount']= df1['Quantity'] * df1['Price']
```

```python
df1['calc_amount'].sum()
```

```python
revenue_by_product = df1.groupby('Product Class')['calc_amount'].sum().sort_values(ascending=False)
pd.set_option('display.float_format', '{:,.2f}'.format)
revenue_by_product
```

```python
top_managers= df1.groupby('Manager')['calc_amount'].sum().sort_values(ascending=False)
top_managers
```

```python
df[df['Quantity'] < 0]["Manager"].value_counts()
```

```python
top_sales_reps = df1.groupby('Name of Sales Rep')['calc_amount'].sum().sort_values(ascending=False)
top_sales_reps
```

```python
Top_Selling_Cities = df1.groupby('City')['calc_amount'].sum().sort_values(ascending=False)
Top_Selling_Cities
```

```python
top_selling_channel = df1.groupby('Channel')['calc_amount'].sum().sort_values(ascending=False)
top_selling_channel
```

# **Key insights**
**Top-Selling Products**
- Analgesics are the highest revenue-generating category, followed by Antiseptics and Mood Stabilizers.
- Focus marketing efforts and inventory planning on these categories.

**Sales Representatives Performance**

- The top Sales Reps significantly outperformed others in terms of total revenue.
- Reward top performers and provide training/support to lower-performing reps.

**Sales Channels Performance**
- Pharmacies contributed the most to total revenue, followed by Hospitals. Other channels had significantly lower performance.





# **Recommendations**
- Increase promotion of top-selling product categories

- Strengthen partnerships with pharmacies (top-performing sales channel)

- Optimize sales rep assignments based on regional performance

- Invest in training for low-performing Sales Reps and Managers

- Use sales insights to guide pricing, inventory, and expansion decisions
